```python
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.datasets import load_iris
from sklearn.svm import SVC
df=load_iris()
print(dir(df))
```

```python
x=df.data
y=df.target
```

```python
d={
    "kernel":["linear","poly"],
     "C":[20,30,40,50,60]
}
```

```python
model=SVC()
obj=GridSearchCV(model,d,cv=10)
obj.fit(x,y)
obj.cv_results_
df=pd.DataFrame(obj.cv_results_)
df
```

```python
df.columns
```

```python
cols=["param_C","param_kernel","mean_test_score"]
df=df[cols]
df
```

```python
obj.best_params_
```

```python
obj.best_score_
```

```python
import pandas as pd
from sklearn.model_selection import RandomizedSearchCV
from sklearn.datasets import load_iris
from sklearn.svm import SVC
df=load_iris()
print(dir(df))
```

```python
x=df.data
y=df.target
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
df=pd.read_csv("poly_dataset.csv")
df
```

```python
x=df.iloc[:,1:2].values
x
```

```python
y=df.iloc[:,2].values
y
```

```python
model=LinearRegression()
model.fit(x,y)
plt.scatter(x,y)
plt.plot(x,model.predict(x))
```

```python
model=PolynomialFeatures(degree=4)
x_poly=model.fit_transform(x)
x_poly
```

```python
obj=LinearRegression()
obj.fit(x_poly,y)
```

```python
plt.scatter(x,y)
plt.plot(x,obj.predict(x_poly),color="green")
```

```python
result=obj.predict(model.fit_transform(x,y))
result
```

```python
df["prediction"]=result
df
```

```python
import pickle 
class Rajesh:
     def __init__(self,a,b,c):
         self.a=a
         self.b=b
         self.c=c
     def name(self):
         print("first parameter",self.a)
         print("second parameter",self.b)
         print("third parameter",self.c)
with open("rr.pkl","wb")as f1:
 x=Rajesh(100,200,300)
 pickle.dump(x,f1)
 print("pickling is done")
```

```python
 with open("rr.pkl","rb")as f2:
   obj=pickle.load(f2)
   obj.name()
   print("unpickling is done")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
df=pd.read_csv("homeprices2.csv")
df
```
